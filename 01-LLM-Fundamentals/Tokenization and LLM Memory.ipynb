#Tokenization and LLM Memory

This notebook explores two core concepts behind LLM applications:

Tokenization — LLMs process text by breaking it into smaller units called tokens, converting those tokens into numerical IDs, and using those IDs during inference.

Stateless conversations — LLM APIs do not retain conversation history between requests. Each API call is independent, so applications must include prior messages in the request to create conversational continuity.

This notebook demonstrates how text is converted into tokens, how token IDs can be inspected and decoded, how messages are structured for the OpenAI client, and how chat applications manage conversation history.

```python
'''Tokenization setup block - goal is to import a tokenizer library, chooses the tokenizer rules for a specific model, then convert them to a sentence into token IDs
Note: am not calling openai or sending request to an LLM. we only want to see how the text gets prepared for an llm '''

import tiktoken                                                       # contains tools for converting texts in to tokens 

encoding = tiktoken.encoding_for_model("gpt-4.1-mini")                #creating an encoding object and storing it in the variable encoding 

tokens = encoding.encode("this is just a test to see how tokenization works in real time")
```

```python
#words turned in to a list of tokens(vectors)
tokens
```

```python
#a for loop going through each token ID and converting it back into text 
for token_id in tokens:
    token_text = encoding.decode([token_id])
    print(f"{token_id} = {token_text}")
```

```python
#Decoding one token ID back into readable text
encoding.decode([2860])
```

```python
#confirming whether api key is available before i make an api request 
import os                           #library that lets python read environment variables 
from dotenv import load_dotenv      #used to load values from a .env into environment 

load_dotenv(override=True)                            #loads values from your .env file
api_key = os.getenv('OPENAI_API_KEY')                 #gets the value stored under the name OPEN_API_KEY

if not api_key:
    print("No API key was found")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj")
else:
    print("API key found and looks good so far!")
```

```python
from openai import OpenAI      #import OpenAI class from the openai library 
openai = OpenAI()              #creating the client object and storing it 
```

```python
#give the model the rules for how the assistant should behave and the message from the user
# Every time you send a message, the app creates a new request to the LLM.
#The LLM itself does not automatically remember the last request.
#The app manages the conversation context and sends enough earlier information again so the LLM can respond correctly

messages = [
    {"role": "system", "content": "You are a helpful assistant with 30 year of experience"},
    {"role": "user", "content": "this is a test"}

    ]
```

```python
#sending the messages list to gpt model and storing the full reply object in reponse
response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
response.choices[0].message.content   # pulling out the actual text the model wrote 
```
