```python
import os #lets my program interact with the operatimns system.
from dotenv import load_dotenv #imports a function that read values from a .env file and places them into the program

#note - this pattern is what we constantly use in Ai engineering for openai, anthropic, databases, aws, hugging face and other service 
load_dotenv(override=True)     #loads the values from my .env file into the program's environment
api_key = os.getenv('OPENAI_API_KEY') # Gets the value stored under "OPENAI_API_KEY" and saves it in variable api_key 

if not api_key:
    print("No API key was found") # if not api_key: raise ValueError("OPENAI_API_KEY is missing. Add it to your .env file.") this can also stop the program immediately because the program cannot continue without the key 

elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key") #this checks whether the key begins with sk-proj 
else:
    print("API key found and looks good so far!")
```
