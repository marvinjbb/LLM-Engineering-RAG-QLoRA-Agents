# Calling LLMs with APIs

In this notebook, I am seeing how Python communicates with Large Language Models through APIs.

I will load my OpenAI API key securely from a `.env` file, send requests to an LLM, and receive AI-generated responses back into Python.
i'm 
* Using `requests` to see the raw API request and response process.
* Using the OpenAI Python library, which is the cleaner and more common approach.

The main goal is to understand the standard pattern for connecting Python applications to AI models so I can later use OpenAI, Gemini, Ollama, and other LLM providers in real projects.


```python
import os #lets my program interact with the operatimns system.
from dotenv import load_dotenv #imports a function that read values from a .env file and places them into the program

#note - this pattern is what we constantly use in Ai engineering for openai, anthropic, databases, aws, hugging face and other service 
load_dotenv(override=True)     #loads the values from my .env file into the program's environment
api_key = os.getenv('OPENAI_API_KEY') # Gets the value stored under "OPENAI_API_KEY" and saves it in variable api_key 

if not api_key:
    print("No API key was found") # if not api_key: raise ValueError("OPENAI_API_KEY is missing. Add it to your .env file.") this can also stop the program immediately because the program cannot continue without the key 

elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key") #this checks whether the key begins with sk-proj 
else:
    print("API key found and looks good so far!")
```

```python
import requests  # library that lets python send HTTP requests to websites and APIs 
#reponse = requests.post(url, headers= headers, json=payload) standard pattern when calling APIs manually

headers = {"Authorization": f"Bearer {api_key}", # this sends the api key to prove that python program is allowed to use the api
            "Content-Type": "application/json"}  # this tells the api that the data i am send is JSON. this is the standard for apis 

#payload - what u want the model to do. its the actual data sent in the request body
#note payload is a standard API word. means that the data we are sending to the API 
payload = {
    "model": "gpt-5-nano", #the model required label the API expects. gpt 5 is the specific model you chose
    "messages": [ {"role": "user", "content": "Tell me a fun fact"}] #messages a list thats called- the api expects conversation messages in list because a conversation can have more than one message
}

payload
```

```python
#we now have the headers - the api key and json format and the payload that gives us the model name and prompt this is 
#where we send both to openai 
#note: post request sends data to a server and asks it to do something with the data 

response = requests.post( 
    "https://api.openai.com/v1/chat/completions",
    headers=headers,
    json=payload
)

response.json() # this takes the api response, that arrives in josn and converts it into python data.coming a dictionary 

#requests.post() sends a POST request.
#response = saves whatever comes back from the API into the variable response.
#"https://api.openai.com/v1/chat/completions" - this endpoint(the url(the address))- the sqecific url where an API feature lives '''
```

```python
#this line pulls out only the ai text answer from the large api response
response.json()["choices"][0]["message"]["content"]

#note this is a better to do this.this is where openai packages comes in so we dont have to write this all the time 
```

```python
#Note: with requests we had to manually write everything. the headers payload and response- its only used to understand whats happening underneath
#doing it manually everytime is annoying and easier to messup. so openai provides a python library that wraps that work for us

from openai import OpenAI  #imports the openAI client class from the openAI library. A class is a blueprint for creating objects that have useful capabilites
client = OpenAI() # create an openai client object - connection tool for talking to openai. 

#this says Use this OpenAI client, go to the chat completions feature, and create a new completion.
response = client.chat.completions.create(model="gpt-5-nano", messages=[{"role": "user", "content": "what's considered to be the best ai today"}])

response.choices[0].message.content


#Note: Because you already loaded OPENAI_API_KEY into your environment using .env, OpenAI() can usually find it automatically.
```

```python
# a more modern way to do this 
from openai import OpenAI

client = OpenAI()

response = client.responses.create(
    model = "gpt-5-nano", 
    input = "list the wonders of the world") 
print (response.output_text)


```

```python
#can use olluma. Ollama runs locally at http://localhost:11434 by default, and local requests do not need an API key

from ollama import chat

response = chat(
    model="llama3.2",
    messages=[
        {"role": "user", "content": "Tell me a fun fact"}
    ]
)

print(response["message"]["content"])


#note: ollama is software that lets you download and run AI models on your own computer instead of sending every prompt to openai, gemini or anthropic
```
